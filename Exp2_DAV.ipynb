# **Implementation in Python**

```python
!pip install pandas numpy matplotlib seaborn scikit-learn
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

df = pd.read_csv('company.csv')

X = df['TV']
y = df['Sales']

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

coefficient_of_determination = model.rsquared
intercept = model.params['const']
coefficients = model.params['TV']

print(f'Coefficient of Determination (R^2): {coefficient_of_determination:.4f}')
print(f'Intercept: {intercept:.4f}')
print(f'Coefficient for TV: {coefficients:.4f}')

predictions = model.predict(X)

plt.scatter(df['TV'], df['Sales'], label='Actual Data')
plt.plot(df['TV'], predictions, label='Linear Regression', color='red')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Linear Regression Model')
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(df['TV'], bins=20, color='blue', edgecolor='black')
plt.title('Histogram of TV Advertising Budget')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Frequency')
plt.show()
```
